```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']

# A. Una función que calcule el precio total de la compra

def precio_total_compra(articulos,compra):
    precio = 0
    for index in compra:
        precio += articulos[index]
    return precio

print(precio_total_compra(articulos,compra))

# B. Una función que calcule el precio total de la compra con un descuento del 10%

def precio_total_desc(articulos,compra):
    precio = precio_total_compra(articulos,compra)
    return precio*0.9

print(precio_total_desc(articulos, compra))

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def precio_desc_compraSup100(articulos,compra):
    precio = precio_total_compra(articulos,compra)  
    if precio > 100:
        precio = precio_total_desc(articulos,compra)
    return precio

print(precio_desc_compraSup100(articulos,compra))

# D. Una función que calcule el precio total aplicando el IVA (21%)

def precio_IVA(articulos,compra):
   precio = precio_total_compra(articulos,compra) * 1.21
   return round(precio,2)

print(precio_IVA(articulos,compra))

# E.lista los artículos cuyo precio es superior a 20€

def art_precioSup20(articulos):
  sup20 = []
  for key, value in articulos.items():
    if value > 20:
      sup20.append(key)
  return sup20

art_precioSup20(articulos)
```
